Datetime format - Extract the months from the dataset and store in a separate column. Then filter the data to show only the information for the first quarter , ie. January, February and March. Hint: If data from March does not exist, consider only January and February.

```python

import datetime
```

```python
file['effective to date']
```

```python
file['effective to date'] = pd.to_datetime(file['effective to date'], errors='coerce')
file['effective to date']
```

```python
file['month'] = pd.DatetimeIndex(file['effective to date']).month

file
```

```python
file= file[file['month'] <= 3]
```

```python
file
```

```python
# saving the dataframe
file.to_csv('finalfile.csv', index=False)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
pd.set_option('display.max_columns', None) 

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('marketing_customer_analysis.csv')
```

## Show DataFrame info.

```python
data.info()
```

## Describe DataFrame.

```python
data.describe()
```

## Show a plot of the total number of responses.

```python
sns.countplot('Response', data=data)
```

## Show a plot of the response rate by the sales channel.

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

sns.histplot(binwidth=0.5, x="Response", hue="Sales Channel", data=data, stat="count", multiple="stack")
```

## Show a plot of the response rate by the total claim amount

```python
plt.figure(figsize=(16,8)) 

sns.barplot(x="Response", y="Total Claim Amount", data=data) 
plt.show()
```

# Show a plot of the response rate by income.

```python
plt.figure(figsize=(12,8)) 

sns.barplot(y="Response", x="Income", data=data) 
plt.show()
```
